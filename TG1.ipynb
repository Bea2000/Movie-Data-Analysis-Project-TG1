```
#https://stackoverflow.com/questions/30462807/encoding-error-in-panda-read-csv (por error= "'utf-8' codec can't decode byte...")
import pandas as pd
import altair as alt
df = pd.read_csv("movies.csv", encoding="latin-1")
```

```
#Pregunta 1

#Primero se redondean los puntajes (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.round.html)
scoresround=df.round(0)
scoresround
```

```
#https://www.firstpythonnotebook.org/groupby/ (aprender a usar el groupby)
#https://riptutorial.com/es/pandas/example/5744/renombrar-una-columna (cambiarle el nombre a las columnas)
a=scoresround.groupby('score').count()[['name']].reset_index() #organizamos nuestra tabla por puntajes obtenidos y contamos la cantidad de películas con esos puntajes
a= a.rename(columns={"name":"Cantidad de películas","score":"Puntaje"})
a
```

```
#https://altair-viz.github.io/user_guide/configuration.html (cómo agregar título al gráfico)
#https://altair-viz.github.io/gallery/simple_bar_chart.html (como crear gráfico de barras)
alt.Chart(a).mark_bar().encode(
    x='Puntaje',
    y='Cantidad de películas',
).interactive().properties(
    title='Gráfico 1: Distribución de puntajes'
)
```

```
#Pregunta 2

#https://stackoverflow.com/questions/43859416/finding-top-10-in-a-dataframe-in-pandas (función nlargest para conseguir los puntajes más altos)
topten=df.nlargest(10, 'score')

topten
```

```
#Pregunta 3

#https://codereview.stackexchange.com/questions/185389/dropping-rows-from-a-pandas-dataframe-where-some-of-the-columns-have-value-0/185390 (eliminar los valores=0)
one=df.copy()
drop_A=df.index[one["budget"] == 0].tolist()
one=one.drop(one.index[drop_A])
one= one.rename(columns={"score":"Puntaje","budget":"Presupuesto","rating":"Rating"})
one
```

```
#https://altair-viz.github.io/gallery/scatter_tooltips.html (Como hacer un diagrama de dispersión)
alt.Chart(one).mark_circle(size=20).encode(
    x='Presupuesto',
    y='Puntaje',
    color='Rating',
    tooltip=['name', 'Rating', 'Presupuesto', 'Puntaje']
).interactive().properties(
    title='Gráfico 3: ¿Importa el presupuesto?' #Puse un nombre distinto al del ejemplo porque o si no, no tiene mucho sentido.
)

#A partir del gráfico diría que el presupuesto es de cierta forma determinante del puntaje ya que notamos como péliculas con alto prespuesto solo tienen puntajes relativamente altos. Aún así vemos como varias películas con bajos presupuestos tienen altos puntajes, por lo que no podríamos asegurar con certeza la proporción directa en presupuesto y puntaje.
```

```
 #Pregunta 4

 #https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html
copia=df.copy()
copia= copia.drop(copia[copia['released'].map(len) == 4].index)
copia['released']=pd.to_datetime(copia['released'])
copia
```

```
#https://www.w3schools.com/python/python_datetime.asp (convertir los meses a números)
#https://stackoverflow.com/questions/44908383/how-can-i-group-by-month-from-a-date-field-using-python-pandas (como agrupar por meses)
tabla=copia.groupby(copia['released'].dt.strftime('%m'))['score'].mean().sort_values().reset_index()
tabla= tabla.rename(columns={"score":"Puntaje","released":"Mes"})
tabla
```

```
alt.Chart(tabla).mark_bar().encode(
   alt.X('Mes'),
   alt.Y('Puntaje'),
).interactive().properties(
    title='Gráfico 4: ¿Importa el mes?'
)
```
